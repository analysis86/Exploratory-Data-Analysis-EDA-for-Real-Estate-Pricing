# 3. Univariate Analysis:

# Import libraries:

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_data.head()
```

```python
housing_data.tail()
```

```python
housing_data.shape
```

# Numeric variables:

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
numeric_vars = ['LotFrontage', 'LotArea', 'MasVnrArea', '1stFlrSF', 'GrLivArea']
for var in numeric_vars:
    plt.figure(figsize=(8, 6))
    sns.histplot(data=housing_data, x=var, kde=True)
    plt.title(f'Distribution of {var}')
    plt.show()
```

```python
categorical_vars = ['Identifies_the_type_of_dwelling', 'Street', 'Lot_configuration', 'LotShape']
for var in categorical_vars:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=housing_data, x=var)
    plt.title(f'Count of {var}')
    plt.show()
```

# Sale-price distribution:

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=housing_data, x='SalePrice', kde=True)
plt.title('Distribution of SalePrice')
plt.show()
```

# Histogram:

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=housing_data, x='SalePrice', bins=30, kde=True)
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.title('Histogram of Sale Price')
plt.show()
```

# Box-plot:

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=housing_data, y='SalePrice')
plt.ylabel('Sale Price')
plt.title('Box Plot of Sale Price')
plt.show()
```

# Kernel Density Estimation:

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=housing_data['SalePrice'], shade=True)
plt.xlabel('Sale Price')
plt.ylabel('Density')
plt.title('Kernel Density Estimation of Sale Price')
plt.show()
```

# Violin-plot:

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=housing_data, y='OverallQual')
plt.ylabel('Overall Quality')
plt.title('Violin Plot of Overall Quality')
plt.show()
```
